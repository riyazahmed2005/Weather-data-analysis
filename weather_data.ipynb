```python
import plotly.express as px
f=px.line(df,x='Date/Time',y='Temp_C')
f.show()
```

```python
highest_temp=df[df["Temp_C"]==df.loc[:,"Temp_C"].max()]
highest_temp
```

```python
lowest_temp=df[df["Temp_C"]==df.loc[:,"Temp_C"].min()]
lowest_temp
```

```python
fig = px.histogram(df, x='Temp_C', nbins=100, title='Temperature Distribution')
fig.update_layout(xaxis_title='Temperature (°C)', yaxis_title='Frequency')
fig.show()
```

```python
fig = px.box(df, y="Temp_C", title="Temperature Spread and Outliers")
fig.update_layout(yaxis_title="Temperature (°C)")
fig.show()
```
